```python
import eleanor
```

```python
star = eleanor.Source(tic=377780790)
```

```python
print(star.coords)
```

```python
data = eleanor.TargetData(star, height=21, width=21, bkg_size=31, do_psf=True, do_pca=True)
```

```python
import lightkurve as lk
```

```python
kepler10 = data.to_lightkurve()
kepler10.plot()
```

```python
import matplotlib.pyplot as plt
import numpy as np

q = data.quality == 0
plt.figure(figsize=(10,5))
plt.plot(data.time[q], data.raw_flux[q]/np.nanmedian(data.raw_flux[q])+0.06, 'k')
plt.plot(data.time[q], data.corr_flux[q]/np.nanmedian(data.corr_flux[q]) + 0.03, 'r')
plt.plot(data.time[q], data.pca_flux[q]/np.nanmedian(data.pca_flux[q]), 'g')
plt.plot(data.time[q], data.psf_flux[q]/np.nanmedian(data.psf_flux[q]) - 0.02, 'b')
plt.ylabel('Normalized Flux')
plt.xlabel('Time [BJD - 2457000]')
```

```python
def plot_eleanor_lightcurves(ticid):
    star = eleanor.Source(tic = ticid)
    data = eleanor.TargetData(star, height=21, width=21, bkg_size=31, do_psf=True, do_pca=True)
    q = data.quality == 0
    plt.figure(figsize=(14,8))
    plt.plot(data.time[q], data.raw_flux[q]/np.nanmedian(data.raw_flux[q])+0.06, 'k.')
    plt.plot(data.time[q], data.corr_flux[q]/np.nanmedian(data.corr_flux[q]) + 0.03, 'r.')
    plt.plot(data.time[q], data.pca_flux[q]/np.nanmedian(data.pca_flux[q]), 'g.')
    plt.plot(data.time[q], data.psf_flux[q]/np.nanmedian(data.psf_flux[q]) - 0.02, 'b.')
    plt.ylabel('Normalized Flux')
    plt.xlabel('Time [BJD - 2457000]')
```

```python
plot_eleanor_lightcurves(350814622)
```
